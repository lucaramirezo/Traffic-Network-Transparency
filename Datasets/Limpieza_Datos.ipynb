## Accidentalidad 2022-2023

El fichero 202X_Accidentalidad.csv recoge los datos de Accidentes de tráfico en la Ciudad de Madrid registrados por la Policía Municipal. 

1. Tomamos los datos de los ficheros 2022_Accidentalidad.csv y 2023_Accidentalidad.csv y creamos dataframes con los datos de los ficheros

```python
accidentes_2022 = pd.read_csv('./2022_Accidentalidad.csv', sep = ';')
accidentes_2022
```

```python
accidentes_2023 = pd.read_csv('./2023_Accidentalidad.csv', sep = ';')
accidentes_2023
```

2. Eliminamos las columnas 'sexo', cod_lesividad', y 'lesividad' de los dataframes y unas columnas de error que hay en 2022, ya que no nos interesan para este estudio

```python
accidentes_2022.drop(columns=['sexo', 'cod_lesividad', 'lesividad', 'Unnamed: 19', 'Unnamed: 20'], inplace=True)
accidentes_2023.drop(columns=['sexo', 'cod_lesividad', 'lesividad'], inplace=True)
```

3. Si hay filas duplicadas en los dataframes las eliminamos para evitar tener datos repetidos

```python
accidentes_2022.drop_duplicates(inplace=True)
accidentes_2023.drop_duplicates(inplace=True)
```

4. Renombramos las columnas 'estado_meteorológico' para evitar acentos y mantener la uniformidad

```python
accidentes_2022.rename(columns={'estado_meteorológico': 'estado_meteorologico'}, inplace=True)
accidentes_2023.rename(columns={'estado_meteorológico': 'estado_meteorologico'}, inplace=True)
```

5. Comparamos si las columnas de los dos dataframes son iguales para verificar su uniformidad

```python
accidentes_2022.columns.equals(accidentes_2023.columns) 
```
